````python
# Credit Scoring - Google Colab ready notebook
# Purpose: EDA, preprocessing, modeling, evaluation and markdown report generation
# Instructions:
# 1) Upload your dataset (CSV) to Colab via the left Files panel or mount Google Drive and update 'DATA_PATH'.
# 2) Run each cell sequentially. The script will save a Markdown report and trained model artifacts.

# -------------------------
# Cell 0: Install dependencies (Colab)
# -------------------------
!pip install -q scikit-learn==1.3.2 xgboost==3.1.2 imbalanced-learn==0.11.0 pandas-profiling==3.6.5 matplotlib==3.7.2 seaborn==0.12.2

# -------------------------
# Cell 1: Imports
# -------------------------
import os
import io
import json
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve, auc, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.calibration import calibration_curve
from sklearn.inspection import permutation_importance

# -------------------------
# Cell 2: Configuration
# -------------------------
# If you uploaded dataset via Colab Files: set DATA_PATH to the filename (e.g. '/content/credit_data.csv')
# If you mounted Google Drive, set DATA_PATH accordingly (e.g. '/content/drive/MyDrive/credit_data.csv')
DATA_PATH = '/content/credit_data.csv'  # <-- update if needed
TARGET_COL = 'target'  # name of binary target column: 1 = default (bad), 0 = good. Update if different.
RANDOM_STATE = 42
OUTPUT_DIR = '/content/credit_scoring_output'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Create a dummy CSV for demonstration if not found
if not os.path.exists(DATA_PATH):
    print(f"{DATA_PATH} not found. Creating a dummy dataset for demonstration.")
    np.random.seed(RANDOM_STATE)
    data = {
        'loan_amount': np.random.randint(1000, 50000, 100),
        'age': np.random.randint(20, 70, 100),
        'income': np.random.randint(15000, 100000, 100),
        'education': np.random.choice(['high_school', 'bachelor', 'master', 'phd'], 100),
        'employment_years': np.random.randint(0, 40, 100),
        'num_dependents': np.random.randint(0, 5, 100),
        'credit_score': np.random.randint(300, 850, 100),
        'loan_purpose': np.random.choice(['debt_consolidation', 'home_improvement', 'car', 'other'], 100),
        'region': np.random.choice(['north', 'south', 'east', 'west'], 100),
        'has_housing': np.random.choice([0, 1], 100),
        'target': np.random.choice([0, 1], 100, p=[0.8, 0.2]) # 20% default rate
    }
    dummy_df = pd.DataFrame(data)
    dummy_df.to_csv(DATA_PATH, index=False)
    print(f"Dummy data saved to {DATA_PATH}")

# -------------------------
# Cell 3: Load data (robust)
# -------------------------
# Data loading will now always succeed because a dummy file is created if DATA_PATH doesn't exist.
df = pd.read_csv(DATA_PATH)
print('Data loaded:', DATA_PATH)
print('Shape:', df.shape)
display(df.head())

# -------------------------
# Cell 4: Quick EDA
# -------------------------
# Basic info
try:
    display(df.info())
    display(df.describe(include='all').T)
except NameError:
    print('Load the data first (run the load cell).') # This should no longer be reached

# Missing values
if 'df' in globals():
    mv = df.isnull().sum().sort_values(ascending=False)
    display(mv[mv>0])

    # Plot target balance if present
    if TARGET_COL in df.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(x=TARGET_COL, data=df)
        plt.title('Target distribution')
        plt.savefig(os.path.join(OUTPUT_DIR,'target_distribution.png'))
        plt.show()

# -------------------------
# Cell 5: Feature types detection & simple cleaning
# -------------------------
# Auto-detect numeric and categorical columns (exclude target)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
if TARGET_COL in num_cols:
    num_cols.remove(TARGET_COL)

print('Numeric cols:', len(num_cols))
print('Categorical cols:', len(cat_cols))

# Example: drop identifier columns if present
possible_id_cols = [c for c in df.columns if c.lower().startswith('id') or c.lower().endswith('id')]
print('Possible identifier columns to drop:', possible_id_cols)

# -------------------------
# Cell 6: Train/test split
# -------------------------
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=RANDOM_STATE)
print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)

# -------------------------
# Cell 7: Preprocessing pipelines
# -------------------------
# Numeric pipeline: impute (median) + scale
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
# Categorical pipeline: impute (most_frequent) + onehot (drop='if_binary' to reduce dim)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, num_cols),
    ('cat', categorical_transformer, cat_cols)
], remainder='drop')

# -------------------------
# Cell 8: Handle imbalance (SMOTE) inside pipeline via custom fit
# -------------------------
# We'll build models separately and apply SMOTE on training set after preprocessing.
X_train_pre = preprocessor.fit_transform(X_train)
X_test_pre = preprocessor.transform(X_test)

# Keep feature names for importance plots
def get_feature_names(preprocessor):
    names = []
    if 'num' in preprocessor.named_transformers_:
        names += num_cols
    if 'cat' in preprocessor.named_transformers_:
        ohe = preprocessor.named_transformers_['cat'].named_steps['onehot']
        cat_names = ohe.get_feature_names_out(cat_cols).tolist()
        names += cat_names
    return names

feature_names = get_feature_names(preprocessor)
print('Total features after preprocessing:', len(feature_names))

# Apply SMOTE
sm = SMOTE(random_state=RANDOM_STATE)
X_train_bal, y_train_bal = sm.fit_resample(X_train_pre, y_train)
print('After SMOTE train shape:', X_train_bal.shape)

# -------------------------
# Cell 9: Model training and evaluation helpers
# -------------------------
models = {
    'logistic': LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    'random_forest': RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1),
    'xgboost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=RANDOM_STATE, n_jobs=-1)
}

results = {}

def evaluate_model(clf, Xtr, ytr, Xte, yte, name):
    clf.fit(Xtr, ytr)
    probs = clf.predict_proba(Xte)[:,1]
    preds = clf.predict(Xte)
    roc = roc_auc_score(yte, probs)
    fpr, tpr, _ = roc_curve(yte, probs)
    precision, recall, _ = precision_recall_curve(yte, probs)
    pr_auc = auc(recall, precision)
    cm = confusion_matrix(yte, preds)
    report = classification_report(yte, preds, digits=4)
    results[name] = {'model': clf, 'roc_auc': roc, 'fpr': fpr, 'tpr': tpr, 'precision': precision, 'recall': recall, 'pr_auc': pr_auc, 'confusion_matrix': cm, 'report': report}
    print(f"{name} ROC AUC: {roc:.4f} | PR AUC: {pr_auc:.4f}")
    print(report)

# -------------------------
# Cell 10: Train baseline models on balanced training set
# -------------------------
for name, model in models.items():
    print('Training', name)
    evaluate_model(model, X_train_bal, y_train_bal, X_test_pre, y_test, name)

# -------------------------
# Cell 11: Plot ROC and PR curves
# -------------------------
plt.figure(figsize=(8,6))
for name in results:
    plt.plot(results[name]['fpr'], results[name]['tpr'], label=f"{name} (AUC={results[name]['roc_auc']:.3f})")
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR,'roc_curves.png'))
plt.show()

plt.figure(figsize=(8,6))
for name in results:
    plt.plot(results[name]['recall'], results[name]['precision'], label=f"{name} (PR-AUC={results[name]['pr_auc']:.3f})")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR,'pr_curves.png'))
plt.show()

# -------------------------
# Cell 12: Feature importance (for tree-based models) + permutation importance for logistic
# -------------------------
best_tree = results['random_forest']['model']
importances = None
if hasattr(best_tree, 'feature_importances_'):
    importances = best_tree.feature_importances_
    fi = pd.Series(importances, index=feature_names).sort_values(ascending=False).head(30)
    plt.figure(figsize=(8,10))
    fi.plot(kind='barh')
    plt.gca().invert_yaxis()
    plt.title('Random Forest - Top 30 feature importances')
    plt.savefig(os.path.join(OUTPUT_DIR,'rf_feature_importance.png'))
    plt.show()

# Permutation importance for logistic
from sklearn.inspection import permutation_importance
log_model = results['logistic']['model']
perm = permutation_importance(log_model, X_test_pre, y_test, n_repeats=10, random_state=RANDOM_STATE, n_jobs=-1)
perm_imp = pd.Series(perm.importances_mean, index=feature_names).sort_values(ascending=False).head(30)
plt.figure(figsize=(8,10))
perm_imp.plot(kind='barh')
plt.gca().invert_yaxis()
plt.title('Logistic - Permutation importance (top 30)')
plt.savefig(os.path.join(OUTPUT_DIR,'logistic_permutation_importance.png'))
plt.show()

# -------------------------
# Cell 13: Calibration plot
# -------------------------
plt.figure(figsize=(8,6))
for name in results:
    prob_pos = results[name]['model'].predict_proba(X_test_pre)[:,1]
    frac_pos, mean_pred = calibration_curve(y_test, prob_pos, n_bins=10)
    plt.plot(mean_pred, frac_pos, marker='o', label=name)
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration plots')
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR,'calibration.png'))
plt.show()

# -------------------------
# Cell 14: Save best model (by ROC AUC)
# -------------------------
best_name = max(results.keys(), key=lambda k: results[k]['roc_auc'])
best_model = results[best_name]['model']
import joblib
joblib.dump(best_model, os.path.join(OUTPUT_DIR, f'best_model_{best_name}.pkl'))
print('Best model:', best_name, 'saved to output dir')

# -------------------------
# Cell 15: Generate Markdown report and save
# -------------------------
REPORT_MD = f"""
# Credit Scoring - Analyse automatique

**Date générée:** {datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S UTC')}

## 1. Introduction
Ce rapport présente une analyse complète (EDA, prétraitement, modélisation et évaluation) d'un jeu de données de crédit.

## 2. Jeu de données
- Fichier utilisé: `{os.path.basename(DATA_PATH)}`
- Dimensions: {df.shape}
- Colonne cible: `{TARGET_COL}` (1 = défaut / mauvais, 0 = bon)

## 3. Nettoyage et préparation
- Détection des colonnes numériques: {len(num_cols)}
- Détection des colonnes catégorielles: {len(cat_cols)}
- Identifier columns potentiels supprimés: {possible_id_cols}
- Méthodes d'imputation: median (numérique), most_frequent (catégorique)
- Encodage: One-Hot pour catégorielles
- Standardisation: StandardScaler pour numériques
- Gestion du déséquilibre: SMOTE appliqué au jeu d'entraînement

## 4. Modèles entrainés
- LogisticRegression
- RandomForestClassifier
- XGBClassifier

## 5. Résultats principaux
"""

# Append model metrics
for name in results:
    REPORT_MD += f"""
### {name}
- ROC AUC: {results[name]['roc_auc']:.4f}
- PR AUC: {results[name]['pr_auc']:.4f}
- Matrice de confusion:
```
{results[name]['confusion_matrix']}
```
- Classification report:
```
{results[name]['report']}
```
"""

REPORT_MD += f"\nBest model selon ROC AUC: **{best_name}**\n"
REPORT_MD += "\n## 6. Visualisations enregistrées\n"
REPORT_MD += "Les fichiers suivants ont été sauvegardés dans le dossier `credit_scoring_output` :\n\n"
for fname in ['target_distribution.png','roc_curves.png','pr_curves.png','rf_feature_importance.png','logistic_permutation_importance.png','calibration.png']:
    REPORT_MD += f"- {fname}\n"

REPORT_MD += "\n## 7. Recommandations opérationnelles\n"
REPORT_MD += "- Valider la qualité des données manquantes (source et raison).\n- Tester la calibration (recalibration si nécessaire, p.ex. Platt scaling / isotonic).\n- Mettre en place un pipeline de scoring en production avec surveillance du drift.\n- Documenter et versionner les données et modèles.\n"

# Save report
with open(os.path.join(OUTPUT_DIR,'credit_scoring_report.md'), 'w', encoding='utf-8') as f:
    f.write(REPORT_MD)
print('Markdown report saved to', os.path.join(OUTPUT_DIR,'credit_scoring_report.md'))

# Also save the raw report variable as JSON for programmatic access
with open(os.path.join(OUTPUT_DIR,'report_metadata.json'), 'w') as f:
    json.dump({'generated': datetime.utcnow().isoformat(), 'best_model': best_name}, f)

# -------------------------
# Final instructions cell
# -------------------------
print('\n---- FIN ----')
print('1) Le dossier', OUTPUT_DIR, 'contient les figures, le modèle et le rapport markdown.')
print('2) Pour télécharger le rapport depuis Colab: clic droit -> Télécharger sur credit_scoring_report.md')
print('3) Pour pousser sur GitHub: téléchargez le fichier .md et le notebook, puis commit.')

# ------------------------------------------------------------------
# NOTE: The variable REPORT_MD contains the full Markdown report and is saved as credit_scoring_report.md
# You can open and edit it before pushing to GitHub.
# ------------------------------------------------------------------
````
